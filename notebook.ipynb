Common libraries and directories

```python
import polars as pl
from pathlib import Path

root_dir = Path('.').absolute()
data_dir = root_dir / "data"
```

Merging two CSV files

```python
# EXAMPLE OF MERGING 2 CSV FILES
LEFT = data_dir / "pairs.csv"
RIGHT = data_dir / "overview.csv"
OUT = data_dir / "something.csv"
l, r = pl.read_csv(LEFT), pl.read_csv(RIGHT)
r = r.select("graph", "nodes", "edges")
l = l.join(r, on=["graph"], how="inner", validate="1:1")
# l.write_csv(OUT) <- uncomment this line to save csv
```
