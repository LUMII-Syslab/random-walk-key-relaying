Merging two CSV files

```python
def merge(lhs_csv: Path, rhs_csv:Path, save_as:Path, key_cols:list[str], rhs_cols: list[str]):
    l, r = pl.read_csv(lhs_csv), pl.read_csv(rhs_csv).select(key_cols+rhs_cols)
    out = l.join(r, on=key_cols, how="inner", validate="1:1")
    out.write_csv(save_as)

# example usage
# merge(
#     lhs_csv = data_dir / "pairs.csv",
#     rhs_csv = data_dir / "overview.csv",
#     save_as = data_dir / "pairs_merged.csv",
#     key_cols = ["graph"],
#     rhs_cols = ["nodes", "edges"]
# )
```
